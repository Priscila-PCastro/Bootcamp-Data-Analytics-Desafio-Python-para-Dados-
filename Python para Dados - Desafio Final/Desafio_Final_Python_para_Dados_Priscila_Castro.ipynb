2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]


```python
#importando pandas e numpy

import pandas as pd
import numpy as np
```

```python
#mostrando o dataframe
df_saude_do_sono_estilo_vida.head()
```

```python
#criando uma função
def moda(coluna):
  return coluna.mode()
```

```python
#calculando a moda
df_saude_do_sono_estilo_vida.groupby('Profissão')['Duração do sono'].agg(moda)
```

```python
#calculando a média
df_saude_do_sono_estilo_vida.groupby('Profissão')['Duração do sono'].mean()
```

```python
#calculando a media, moda e a mediana

horas_de_sono_por_profissão = df_saude_do_sono_estilo_vida.groupby('Profissão')['Duração do sono'].agg([np.mean, pd.Series.mode, np.median])
horas_de_sono_por_profissão
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
df_saude_do_sono_estilo_vida.head()
```

```python
df_genero_e_pressão = df_saude_do_sono_estilo_vida[['Gênero', 'Pressão sanguínea']]
df_genero_e_pressão
```

```python
# cópia do subconjunto para evitar o aviso de SettingWithCopyWarning

df_genero_e_pressão = df_genero_e_pressão[['Gênero', 'Pressão sanguínea']].copy()
```

```python
# Extraindo apenas os valores numéricos da coluna de pressão sanguínea
df_genero_e_pressão['Pressão sanguínea'] = df_genero_e_pressão['Pressão sanguínea'].str.extract('(\d+)', expand=False)
df_genero_e_pressão['Pressão sanguínea']
```

```python
# Convertendo a coluna de pressão sanguínea para valores numéricos
df_genero_e_pressão['Pressão sanguínea'] = pd.to_numeric(df_genero_e_pressão['Pressão sanguínea'], errors='coerce')
df_genero_e_pressão['Pressão sanguínea']
```

```python
# Calculando a média da pressão sanguínea para homens e mulheres
media_pressao_sanguinea_por_genero = df_genero_e_pressão.groupby('Gênero')['Pressão sanguínea'].mean()
media_pressao_sanguinea_por_genero
```

```python
#calculando maior pressão sanguínea média
media_pressao_sanguinea_por_genero = media_pressao_sanguinea_por_genero[media_pressao_sanguinea_por_genero == media_pressao_sanguinea_por_genero.max()].index[0]
media_pressao_sanguinea_por_genero
```
